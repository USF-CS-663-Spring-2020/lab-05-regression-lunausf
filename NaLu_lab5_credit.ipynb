```python
import pandas as pd

# Read the data
df = pd.read_csv('Credit.csv')
df
```

```python
#plot predictor against target, limit is target and predictor is income

df.plot(kind='scatter', x='Income', y='Limit')
```

```python
import seaborn as sns 
plot = sns.lmplot(x='Income', y='Limit', data=df) 
```

```python
# This data is appropriate the SLR. 
# We can see the error term is randomly distributed, is not exclusively below or high above, 
# error distance does not increase or decrease with change with x
r1 = sns.residplot('Income', 'Limit', data=df)
r1.set(xlabel='Income', ylabel='Residuals')
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression() #fit regression 
lin_reg.fit(df[['Income']], df['Limit'])
lin_reg
```

```python
lin_reg.coef_ #slop
```

```python
lin_reg.intercept_  #intercept
```
